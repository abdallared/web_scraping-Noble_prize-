## scrape data from table within this link :

https://en.wikipedia.org/wiki/List_of_Nobel_laureates

## data 

* year 
* physiscs
* Physiology or Medicine
* Literature    
* Peace
* Economics (The Sveriges Riksbank prize) 

```python
import requests
from bs4 import BeautifulSoup


respons = requests.get("https://en.wikipedia.org/wiki/List_of_Nobel_laureates")
respons

soup = BeautifulSoup(respons.text , "html.parser")
soup

data  = soup.find ("tbody").find_all("tr")
data


year = data[1].find_all("th")[0].text.split()[0]
phy = data[1].find_all("td")[1].text.split("\n")[0]
chem = data[1].find_all("td")[1].text.split("\n")[0]
py_m = data[1].find_all("td")[2].text.split("\n")[0]
lit = data[1].find_all("td")[3].text.split("\n")[0]
peac = data[1].find_all("td")[4].text.split("\n")[0]
eco = data[1].find_all("td")[5].text.split("\n")[0]


with open ("noble_prize.csv" , mode='w'  , encoding="utf-8") as fd  : 
    fd.write("year,Physics,Chemistry,Physiology_or_Medicine,Literature,peace,Economics\n")
    for i in range(1 , 124 , 1) : 
        print(i)
        if i == 40 or i == 41 or i == 42 :
            # Adjust the indentation of the code block inside the if statement
            year = data[i].find_all("th")[0].text.split()[0]
            phy = data[40].find_all("td")[0].text.split("\n")[0]
            chem = data[40].find_all("td")[0].text.split("\n")[0]
            lit = data[40].find_all("td")[0].text.split("\n")[0]
            peac = data[40].find_all("td")[0].text.split("\n")[0]
            if i >= 69 :
                eco = data[i].find_all("td")[5].text.split("\n")[0]
            fd.write(f"\"{year}\",\"{phy}\",\"{chem}\",\"{py_m}\",\"{lit}\",\"{peac}\",\"{eco}\"\n")
        else : 
            year = data[i].find_all("th")[0].text.split()[0]
            phy = data[i].find_all("td")[1].text.split("\n")[0]
            chem = data[i].find_all("td")[1].text.split("\n")[0]
            py_m = data[i].find_all("td")[2].text.split("\n")[0]
            lit = data[i].find_all("td")[3].text.split("\n")[0]
            peac = data[i].find_all("td")[4].text.split("\n")[0]
            if i >= 69 :
                eco = data[i].find_all("td")[5].text.split("\n")[0]
            fd.write(f"\"{year}\",\"{phy}\",\"{chem}\",\"{py_m}\",\"{lit}\",\"{peac}\",\"{eco}\"\n")
            


import pandas as pd
df = pd.read_csv("noble_prize.csv")
df
```
